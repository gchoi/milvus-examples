## Experiment 1
#### 개요
1. db_desc_gen으로 생성한 table description을 불러온다.
   - table description 에는 table comment, column descriptions가 포함된다.
2. Doc 1개에 table comment, column descriptions를 모두 함께 넣는다. 즉, 1개의 테이블당 1개의 doc이 생성된다. 
3. 임베딩 모델은 bge-m3를 사용함

#### 데이터
- 사용한 데이터는 `data/` 폴더 내 `_generated_schemas_shilla_t2s_devloper_gemma3_12b_20250528_shilla_t2s_devloper_gemma3_12b_20250528.json` 이다. 
  - 해당 데이터는 gemma3:12b모델을 이용하여 생성함.
  - t2s_developer로 select할 수 있는 7개 테이블이 존재한다.
- 따라서 experiment 1에서는 7개의 docs가 생성되며 이는 1개의 collection에 들어간다.

#### Collection name rule
- `<데이터종류>_<계정>_<모델>_<desc생성날짜>_<ver>` 정도로 하면 어떨까 싶습니다. 의견 부탁드립니다~
- 예] `shilla_t2s_developer_gemma3_12b_20250528_v1`

```python
from langchain_ollama import OllamaEmbeddings

from milvus.preprocess.preprocess import load_td, preprocess_td
from milvus.preprocess.td2docs import td2docs_v1
from milvus.client.create_collection import create_collection
```

```python
json_path = "./data/_generated_schemas_shilla_t2s_devloper_gemma3_12b_20250528_shilla_t2s_devloper_gemma3_12b_20250528.json"
URI = "http://10.128.0.20:19530"
```

```python
# Convert table description to docs
data = load_td(json_path)
td = preprocess_td(data)
docs = td2docs_v1(td)
```

```python
embedding = OllamaEmbeddings(model="bge-m3", base_url="http://10.128.0.20:11434")
create_collection(
    docs=docs,
    collection_name="shilla_t2s_developer_gemma3_12b_20250528_v1",
    embedding=embedding,
    uri=URI,
    overwrite=True
)
```
